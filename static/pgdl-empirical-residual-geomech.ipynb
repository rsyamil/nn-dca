# pgdl - empirical - residual - geomech

Run pgdl-empirical with simulated dataset - Geomechanics

```python
import random
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("paper", rc={"font.size":14,"axes.titlesize":14,"axes.labelsize":14})   

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.cm as cm
from matplotlib.colors import Normalize

import dataloader
import util
import fcnn_emp
import fcnn_residual
```

```python
Bakken = dataloader.DataLoader(dataset_type="GEOMECH", verbose=False)
x_train, x_test, y_train, y_test, _, _ = Bakken.get_data_split()

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

#display the input data
dataloader.histplot(x_train, x_test, name="sim-x-geomech")
```

```python
dataloader.plot_production(y_train, y_test, name="sim-y-geomech")
```

```python
#get and train model
model = fcnn_emp.FCNN(x_train, y_train, n_P=3, name="pgdl-emp-sim-geomech")
model.train(n_batch=10, nb_epoch=500, load=False)
```

```python
#predict Y from X
def predict_by_batch(Nb, N, model, X, timestep=60):
    y_hat = []
    for i in range(int(N/Nb)):
        _ = model.predict(X[i*Nb: i*Nb + Nb])
        y_hat.append(_)
    y_hat = np.reshape(np.array(y_hat), (-1, timestep, 3))
    return y_hat

y_train_hat_f1f2 = predict_by_batch(Nb=10, N=640, model=model.F1F2, X=x_train, timestep=60)
y_test_hat_f1f2 = predict_by_batch(Nb=10, N=160, model=model.F1F2, X=x_test, timestep=60)

print(y_train_hat_f1f2.shape, y_test_hat_f1f2.shape)

util.scatterPlot(y_train.flatten(), y_train_hat_f1f2.flatten(), '$y_{train}$', '$\hat{y}_{train}$', 'b', 'F1F2', ylim=1, name="readme/sim_f1f2_train_emp-geomech")
util.scatterPlot(y_test.flatten(), y_test_hat_f1f2.flatten(), '$y_{test}$', '$\hat{y}_{test}$', 'b', 'F1F2', ylim=1, name="readme/sim_f1f2_test_emp-geomech")
```

```python
#predict P from X
def x2p_predict_by_batch(Nb, N, model, X, p_shape=3):
    p = []
    for i in range(int(N/Nb)):
        _ = model.predict(X[i*Nb: i*Nb + Nb])
        p.append(_)
    p = np.reshape(np.array(p), (-1, p_shape))
    return p

p_train_oil = x2p_predict_by_batch(Nb=10, N=640, model=model.X2P_oil, X=x_train, p_shape=3)
p_test_oil = x2p_predict_by_batch(Nb=10, N=160, model=model.X2P_oil, X=x_test, p_shape=3)

print(p_train_oil.shape, p_test_oil.shape)

p_train_water = x2p_predict_by_batch(Nb=10, N=640, model=model.X2P_water, X=x_train, p_shape=3)
p_test_water = x2p_predict_by_batch(Nb=10, N=160, model=model.X2P_water, X=x_test, p_shape=3)

print(p_train_water.shape, p_test_water.shape)

p_train_gas = x2p_predict_by_batch(Nb=10, N=640, model=model.X2P_gas, X=x_train, p_shape=3)
p_test_gas = x2p_predict_by_batch(Nb=10, N=160, model=model.X2P_gas, X=x_test, p_shape=3)

print(p_train_gas.shape, p_test_gas.shape)

#concatenate the parameters
p_train = np.concatenate((np.expand_dims(p_train_oil, axis=-1), np.expand_dims(p_train_water, axis=-1), np.expand_dims(p_train_gas, axis=-1)), axis=2)
p_test = np.concatenate((np.expand_dims(p_test_oil, axis=-1), np.expand_dims(p_test_water, axis=-1), np.expand_dims(p_test_gas, axis=-1)), axis=2)

print(p_train.shape, p_test.shape)
```

```python
#plot sample #20
for c in [20, 21, 237]:
    util.plot_Y(y_train, y_train_hat_f1f2, p_train, case=c, name="sim-y-yhat-geomech")
```

```python
#plot sample
for c in range(0, 4):
    util.plot_Y(y_test, y_test_hat_f1f2, p_test, case=c, name="sim-y-yhat-geomech")
```

# Residual modeling

```python
print(y_train.shape, y_train_hat_f1f2.shape)
print(y_test.shape, y_test_hat_f1f2.shape)

#calculate residuals
res_train = y_train - y_train_hat_f1f2
res_test = y_test - y_test_hat_f1f2

print(res_train.shape, res_test.shape)
```

```python
#get and train f3 model
model_f3 = fcnn_residual.FCNN(x_train, res_train, name="pgdl-emp-sim-res-geomech")
model_f3.train(n_batch=16, nb_epoch=500, load=False)
```

```python
#scatter plots for the empirical dataset
res_train_hat = model_f3.F3.predict(x_train)
res_test_hat = model_f3.F3.predict(x_test)

print(res_train_hat.shape, res_test_hat.shape)

def RMSE(data1, data2):
    return np.sqrt(np.mean((data1.flatten()-data2.flatten())**2))

#scatter plots for training and testing, color by field label
def scatterPlot_res(data1, data2, xlabel, ylabel, color, title, ylim=0, name=""):
    fig, ax = plt.subplots(figsize=(4, 4))
    plt.scatter(data1, data2, c=color, alpha=0.2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xlim([-0.5, 0.5])
    plt.ylim([-0.5, 0.5])
    plt.title(title)
    ax.annotate(f"RMSE: {RMSE(data1, data2):.3f}", xy=(0.8, 0.8),  xycoords='data',
            xytext=(0.05, 0.9), textcoords='axes fraction',
            horizontalalignment='left', verticalalignment='top')
    print(RMSE(data1, data2))

scatterPlot_res(res_train.flatten(), res_train_hat.flatten(), '$r_{train}$', '$\hat{r}_{train}$', 'b', 'F3', ylim=1, name="readme/sim_r_train_emp-geomech")
scatterPlot_res(res_test.flatten(), res_test_hat.flatten(), '$r_{test}$', '$\hat{r}_{test}$', 'b', 'F3', ylim=1, name="readme/sim_r_test_emp-geomech")
```

```python
#add the residual to the original prediction
y_train_hat_f1f2f3 = res_train_hat + y_train_hat_f1f2
y_test_hat_f1f2f3 = res_test_hat + y_test_hat_f1f2

print(y_train_hat_f1f2f3.shape, y_test_hat_f1f2f3.shape)

util.scatterPlot(y_train.flatten(), y_train_hat_f1f2f3.flatten(), '$y_{train}$', '$\hat{y}_{train}$', 'b', 'F1F2F3', ylim=1, name="readme/sim_f3_train_emp-geomech")
util.scatterPlot(y_test.flatten(), y_test_hat_f1f2f3.flatten(), '$y_{test}$', '$\hat{y}_{test}$', 'b', 'F1F2F3', ylim=1, name="readme/sim_f3_test_emp-geomech")
```
